##Yolov8 Training

```python
%cd {main_dir}

from ultralytics import YOLO

model = YOLO('yolov8s.pt')

results = model.train(
    task='detect',
    data=yaml_file,
    epochs=50,
    imgsz=960,
    plots=True,
    patience=30,
    name='train_8s_50ep_isz960'
)

results_dir = results.save_dir
```
